````python
def convert_to_percentage(score):
    try:
        numerator, denominator = map(float, score.split('/'))
        if denominator == 5:
            return (numerator / denominator) * 100
        elif denominator == 4:
            return (numerator / denominator) * 100
        elif denominator == 10:
            return (numerator / denominator) * 100
        else:
            return None  # Return None for unsupported denominators
    except (ValueError, IndexError):
        return None  # Return None for non-conforming values

# Applying conversion function directly to DataFrame
merged_rot_2['percentage'] = merged_rot_2['originalScore'].apply(convert_to_percentage)

# Filtering out rows with unsupported denominators or non-finite values
valid_rows = merged_rot_2[~merged_rot_2['percentage'].isnull()]

# Rounding the 'percentage' column to the nearest whole number
valid_rows['percentage'] = valid_rows['percentage'].fillna(0)  # Replace NaN with 0
valid_rows['percentage'] = valid_rows['percentage'].round().astype(int)

# Update 'merged_rot_2' with the 'percentage' column
merged_rot_2.update(valid_rows['percentage'])

print(merged_rot_2[['originalScore', 'percentage']])
````

As for explaining what CritiScore is and how it's calculated, here's a Markdown explanation:

---

## Understanding CritiScore

CritiScore is a composite metric that represents an aggregated score based on weighted ratings from various critical sources. It aims to provide a comprehensive evaluation of a movie's reception by considering multiple rating metrics.

### Calculation of CritiScore

1. **Available Metrics**
   - The first step involves assessing the availability of different critical metrics for each movie, such as ratings from different sources like Rotten Tomatoes, IMDb, and audience scores.

2. **Adjusting Weights**
   - Weights are assigned to each available metric based on their presence. If a metric is available for a movie, it contributes to the overall CritiScore calculation; otherwise, it doesn't influence the final score.

3. **Composite Score Calculation**
   - The individual ratings from various sources (such as Rotten Tomatoes, IMDb, etc.) are weighted according to their availability and significance.
   - These weighted ratings are then aggregated to compute the CritiScore for each movie.

4. **Normalization to a 0-100 Scale**
   - To ensure uniformity and comparability across different movies, the CritiScore is scaled to a range between 0 and 100.
   - Scaling involves transforming the calculated CritiScore to fit within this standardized range.

5. **Rounding to the Nearest Integer**
   - Finally, each CritiScore value is rounded to the nearest whole number to maintain clarity and simplicity.

The CritiScore provides a consolidated assessment of a movie's reception, considering multiple critical viewpoints, and presents it in a standardized, easy-to-compare format.

---

````python
# Replace commas and convert to float, handling NaN values
merged_data_movies['imbd_votes'] = merged_data_movies['imbd_votes'].str.replace(',', '').astype(float)

# Convert NaN values to 0 before converting to int
merged_data_movies['imbd_votes'] = merged_data_movies['imbd_votes'].fillna(0).astype(int)
````

````python
import pandas as pd

# Normalize imdb_rating to a 0-100 scale
merged_data_movies['normalized_imdb_rating'] = pd.to_numeric(merged_data_movies['imbd_rating'], errors='coerce') * 10


# Define a function to calculate weighted mean
def weighted_mean_with_votes(ratings, votes):
    if isinstance(votes, (int, float)):
        total_votes = votes
    else:
        total_votes = sum(votes)
    
    if total_votes == 0:
        return 0
    
    if isinstance(ratings, (int, float)):
        return ratings
    
    weighted_sum = sum(r * v for r, v in zip(ratings, votes))
    return weighted_sum / total_votes

# Calculate weighted mean of imdb_rating based on imdb_votes
merged_data_movies['imdb_weighted_mean'] = merged_data_movies.apply(lambda row: weighted_mean_with_votes(
    row['normalized_imdb_rating'], 
    row['imbd_votes']  
) if not pd.isnull(row['normalized_imdb_rating']) else 0, axis=1)
````
